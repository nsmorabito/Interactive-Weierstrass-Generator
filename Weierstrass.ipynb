```python
%matplotlib widget
import numpy as np
import matplotlib.pyplot as plt
from matplotlib.widgets import Slider

# Define the Weierstrass function
def weierstrass_function(x, a, b, n_terms):
    return np.sum([a**n * np.cos(b**n * np.pi * x) for n in range(n_terms)], axis=0)

# Initial parameters
a = 0.5
b_init = 3  # Initial value for b
n_terms = 100
x = np.linspace(-2, 2, 1000)
y = weierstrass_function(x, a, b_init, n_terms)

# Create the figure and the line that we will manipulate
fig, ax = plt.subplots()
line, = plt.plot(x, y, lw=2)
ax.set_xlabel('x')
ax.set_ylabel('f(x)')
plt.title('Interactive Weierstrass Function')

# Adjust the main plot to make room for the slider
plt.subplots_adjust(left=0.1, bottom=0.25)

# Make a horizontal slider to control the frequency.
axcolor = 'lightgoldenrodyellow'
ax_b = plt.axes([0.1, 0.1, 0.8, 0.03], facecolor=axcolor)
b_slider = Slider(
    ax=ax_b,
    label='b value',
    valmin=0,
    valmax=15,
    valinit=b_init,
)

# The function to be called anytime a slider's value changes
def update(val):
    line.set_ydata(weierstrass_function(x, a, b_slider.val, n_terms))
    fig.canvas.draw_idle()

# Register the update function with each slider
b_slider.on_changed(update)

# Show the plot
plt.show()
```
